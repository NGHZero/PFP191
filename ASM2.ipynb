```python
def menu():
    print("===========================================")
    print("1. Nhập thông tin của n cuốn sách của FU")
    print("2. In ra màn hình thông tin vừa nhập")
    print("3. Sắp xếp thông tin giảm dần theo năm xuất bản và hiển thị")
    print("4. Tìm kiếm theo tên sách")
    print("5. Tìm kiếm theo tên tác giả")
    print("6. Thoát")
    print("===========================================")


def nhap_thong_tin():
    n = int(input("Nhập số lượng sách: "))
    with open("FU.txt", "w") as file:
        file.write(str(n) + "\n")
        for _ in range(n):
            ten_sach = input("Tên sách: ")
            ten_tac_gia = input("Tên tác giả: ")
            nha_xuat_ban = input("Nhà xuất bản: ")
            nam_xb = input("Năm xuất bản: ")
            gia_ban = input("Giá bán: ")
            file.write(ten_sach + "\n")
            file.write(ten_tac_gia + "\n")
            file.write(nha_xuat_ban + "\n")
            file.write(nam_xb + "\n")
            file.write(gia_ban + "\n\n")


def in_thong_tin():
    with open("FU.txt", "r") as file:
        data = file.readlines()
        total_books = int(data[0])
        print("Tổng số cuốn sách:", total_books)
        print("Tên sách".ljust(30), "Tên tác giả".ljust(30), "Năm xuất bản".ljust(10), "Giá".ljust(10))
        for i in range(1, len(data), 5):
            print(data[i].strip().ljust(30), data[i+1].strip().ljust(30),
                  data[i+3].strip().ljust(10), data[i+4].strip().ljust(10))


def sap_xep_va_in():
    with open("FU.txt", "r") as file:
        data = file.readlines()[1:]
        sorted_data = sorted(zip(data[::5], data[1::5], data[3::5], data[4::5]), key=lambda x: (-int(x[2]), -float(x[3])))
        with open("FU2024.txt", "w") as sorted_file:
            sorted_file.write(str(len(sorted_data)) + "\n")
            for book in sorted_data:
                sorted_file.write(book[0] + book[1] + book[2] + book[3])
        print("Đã sắp xếp và lưu vào FU2024.txt")
        print("Tổng số cuốn sách:", len(sorted_data))
        print("Tên sách".ljust(30), "Tên tác giả".ljust(30), "Năm xuất bản".ljust(10), "Giá".ljust(10))
        for book in sorted_data:
            print(book[0].strip().ljust(30), book[1].strip().ljust(30),
                  book[2].strip().ljust(10), book[3].strip().ljust(10))


def tim_kiem_theo_ten_sach():
    search_term = input("Nhập tên sách cần tìm kiếm: ")
    found = False
    with open("FU.txt", "r") as file:
        data = file.readlines()[1:]
        for i in range(0, len(data), 5):
            if data[i].strip() == search_term:
                print(data[i].strip(), data[i+1].strip(), data[i+2].strip())
                found = True
    if not found:
        print("Không tìm thấy cuốn sách nào!")


def tim_kiem_theo_ten_tac_gia():
    search_term = input("Nhập tên tác giả cần tìm kiếm: ")
    found = False
    with open("FU.txt", "r") as file:
        data = file.readlines()[1:]
        authors = {}
        for i in range(1, len(data), 5):
            author = data[i].strip()
            if author == search_term:
                found = True
                book_name = data[i-1].strip()
                if author in authors:
                    authors[author].append(book_name)
                else:
                    authors[author] = [book_name]

        if found:
            for author, books in authors.items():
                print(author, ", ".join(books), ",", str(len(books)))
        else:
            print("Không tìm thấy tác giả trên!")


def main():
    while True:
        menu()
        choice = input("Nhập lựa chọn của bạn: ")
        if choice == "1":
            nhap_thong_tin()
        elif choice == "2":
            in_thong_tin()
        elif choice == "3":
            sap_xep_va_in()
        elif choice == "4":
            tim_kiem_theo_ten_sach()
        elif choice == "5":
            tim_kiem_theo_ten_tac_gia()
        elif choice == "6":
            print("Thoát chương trình.")
            break
        else:
            print("Lựa chọn không hợp lệ. Vui lòng chọn lại.")


if __name__ == "__main__":
    main()
```
